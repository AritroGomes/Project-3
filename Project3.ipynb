```python
class Caregiver:
    def __init__(self,name,phone,email,payrate,hours):
        self.name = name
        self.phone = phone
        self.email = email
        self.payrate = payrate
        self.hours = hours
        self.days_to_work = []
    am = {}
    pm = {}

    all_caregivers = []
    
    def set_preferences(self):
        days = ["monday","tuesday","wednesday", "thursday", "friday", "saturday","sunday"]
        for day in days:
            availability_am = input(f"What is your availability for {day} am?: ")
            self.am[day] = availability_am
            availability_pm = input(f"What is your availability for {day} pm?: ")
            self.pm[day] = availability_pm
        all_caregivers.append(self)

# anne = Caregiver("anne","443-891-3933","anne@gmail.com",20,10)
# anne.set_preferences()
# anne.am
# anne.pm

class Caregiver:
    caregivers_availability = {}

    def __init__(self, name, phone, email, payrate, hours):
        self.name = name
        self.phone = phone
        self.email = email
        self.payrate = payrate
        self.hours = hours
        self.am = {}
        self.pm = {}

    def set_preferences(self):
        days = ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]
        for day in days:
            availability_am = input(f"What is your availability for {day} am?: ")
            self.am[day] = availability_am
            availability_pm = input(f"What is your availability for {day} pm?: ")
            self.pm[day] = availability_pm
        Caregiver.caregivers_availability[self.name] = {"am": self.am, "pm": self.pm}


    def weekly_schedule(self):          
            html_schedule = f"""
            <html>
            <head>
                <title>Weekly work Schedule for {self.name}</title>
                <style>
                    table {{
                        border-collapse: collapse;
                        width: 100%;
                        margin: 20px 0;
                    }}
                    th, td {{
                        border: 1px solid black;
                        padding: 10px;
                        text-align: center;
                    }}
                    th {{
                        background-color: #f2f2f2;
                    }}
                    td {{
                        height: 100px;
                        vertical-align: top;
                    }}
                </style>
            </head>
            <body>
                <h1>Weekly availability for {self.name}</h1>
                <table>
                    <tr>
                        <th>Mon</th>
                        <th>Tue</th>
                        <th>Wed</th>
                        <th>Thu</th>
                        <th>Fri</th>
                        <th>Sat</th>
                        <th>Sun</th>
                    </tr>
                    <tr>
                        <td>am: {self.am["monday"]}</td>
                        <td>am: {self.am["tuesday"]} </td>
                        <td>am: {self.am["wednesday"]} </td>
                        <td>am: {self.am["thursday"]}</td>
                        <td>am: {self.am["friday"]} </td>
                        <td>am: {self.am["saturday"]} </td>
                        <td>am: {self.am["sunday"]}</td>
                    </tr>
                    <tr>
                        <td>pm: {self.pm["monday"]}</td>
                        <td>pm: {self.pm["tuesday"]} </td>
                        <td>pm: {self.pm["wednesday"]} </td>
                        <td>pm: {self.pm["thursday"]}</td>
                        <td>pm: {self.pm["friday"]} </td>
                        <td>pm: {self.pm["saturday"]} </td>
                        <td>pm: {self.pm["sunday"]}</td>

                    </tr>

                </table>
    
                    
            """

            with open(f"work_schedule_for {self.name}.html", "w") as file:
                file.write(html_schedule)

            print("generating html file")



    def generate_calendar():
        calendar = {}
        for caregiver, availability in Caregiver.caregivers_availability.items():
            for day, time in availability["am"].items():
                if day not in calendar:
                    calendar[day] = {"am": [], "pm": []}
                if time in ["available", "preferred"]:
                    calendar[day]["am"].append(f"{caregiver} ({time})")
            for day, time in availability["pm"].items():
                if time in ["available", "preferred"]:
                    calendar[day]["pm"].append(f"{caregiver} ({time})")
        return calendar

anne = Caregiver("anne", "443-891-3933", "anne@gmail.com", 20, 10)
anne.set_preferences()
anne.weekly_schedule()
print(anne.all_caregivers())


# john = Caregiver("john", "301-287-5690", "john@gmail.com", 25, 15)
# john.set_preferences()

# consolidated_calendar = Caregiver.generate_calendar()
# print("Consolidated Calendar:")
# for day, times in consolidated_calendar.items():
#     print(f"{day.capitalize()} - AM: {', '.join(times['am'])} | PM: {', '.join(times['pm'])}")
```

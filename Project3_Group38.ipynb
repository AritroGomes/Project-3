```python
#Created by Jincy
# caregiver class
import calendar
class Caregiver:
    caregivers_availability = {}

    def __init__(self, name, phone, email, payrate, hours):
        self.name = name
        self.phone = phone
        self.email = email
        self.payrate = payrate
        self.hours = hours
        self.am = {}
        self.pm = {}
        
# Created by Jincy
    def set_preferences(self):
        days = ["monday", "tuesday", "wednesday", "thursday", "friday", "saturday", "sunday"]
        for day in days:
            # user inserts availability for AM and PM Sunday through Saturday
            availability_am = input(f"What is your availability for {day} am?: ")
            self.am[day] = availability_am
            availability_pm = input(f"What is your availability for {day} pm?: ")
            self.pm[day] = availability_pm
        #created by Aritro
        Caregiver.caregivers_availability[self.name] = {"am": self.am, "pm": self.pm}

#Created by Jincy
    def weekly_schedule(self):          
            schedule = f"""
            <html>
            <head>
                <title>Weekly work Schedule for {self.name}</title>
                <style>
                    table {{
                        border-collapse: collapse;
                        width: 100%;
                        margin: 20px 0;
                    }}
                    th, td {{
                        border: 1px solid black;
                        padding: 10px;
                        text-align: center;
                    }}
                    th {{
                        background-color: #f2f2f2;
                    }}
                    td {{
                        height: 100px;
                        vertical-align: top;
                    }}
                </style>
            </head>
            <body>
                <h1>Weekly availability for {self.name}</h1>
                <table>
                    <tr>
                        <th>Mon</th>
                        <th>Tue</th>
                        <th>Wed</th>
                        <th>Thu</th>
                        <th>Fri</th>
                        <th>Sat</th>
                        <th>Sun</th>
                    </tr>
                    <tr>
                    
                        <td>am: {self.am["monday"]}</td>
                        <td>am: {self.am["tuesday"]} </td>
                        <td>am: {self.am["wednesday"]} </td>
                        <td>am: {self.am["thursday"]}</td>
                        <td>am: {self.am["friday"]} </td>
                        <td>am: {self.am["saturday"]} </td>
                        <td>am: {self.am["sunday"]}</td>
                    </tr>
                    <tr>
                        <td>pm: {self.pm["monday"]}</td>
                        <td>pm: {self.pm["tuesday"]} </td>
                        <td>pm: {self.pm["wednesday"]} </td>
                        <td>pm: {self.pm["thursday"]}</td>
                        <td>pm: {self.pm["friday"]} </td>
                        <td>pm: {self.pm["saturday"]} </td>
                        <td>pm: {self.pm["sunday"]}</td>

                    </tr>

                </table>
    
                    
            """
            # creates weekly availability schedule for each caregiver
            with open(f"Weekly Work Schedule For {self.name}.html", "w") as file:
                file.write(schedule)
            # lets user know that a file has been created
            print("generating work schedule")

        
#Created by Aritro
    def generate_calendar():
        # initializes the calendar dictionary
        calendar = {} 
        for caregiver, availability in Caregiver.caregivers_availability.items():
            #checks AM availability for each day
            for day, time in availability["am"].items():
                if day not in calendar:
                    calendar[day] = {"am": [], "pm": []}
                if time in ["available", "preferred", "unavailable"]:
                    # adds caregiver to AM slot
                    calendar[day]["am"].append(f"{caregiver} ({time})")
            #checks PM availability for each day
            for day, time in availability["pm"].items():
                if time in ["available", "preferred", "unavailable"]:
                    # adds caregiver to PM slot
                    calendar[day]["pm"].append(f"{caregiver} ({time})")
        # makes the comlpeted calender
        return calendar

def generate_big_calendar(caregiver_calendar):
    import datetime

    # gets the calendar for November 2024
    cal = calendar.Calendar()
    month_days = cal.itermonthdates(2024, 11)

    schedule = """
    <html>
    <head>
        <title>Caregiver Schedule for November 2024</title>
        <style>
            table {
                border-collapse: collapse;
                width: 100%;
            }
            th, td {
                border: 1px solid black;
                width: 14.28%;
                vertical-align: top;
                height: 100px;
            }
            th {
                background-color: #f2f2f2;
            }
            .other-month {
                background-color: #e6e6e6;
            }
            .date-number {
                font-weight: bold;
            }
        </style>
    </head>
    <body>
        <h1>Caregiver Schedule for November 2024</h1>
        <table>
            <tr>
                <th>Sunday</th>
                <th>Monday</th>
                <th>Tuesday</th>
                <th>Wednesday</th>
                <th>Thursday</th>
                <th>Friday</th>
                <th>Saturday</th>
            </tr>
    """

    week = []
    for date in month_days:
        if date.weekday() == 6:
            if week:
                schedule += "<tr>" + "".join(week) + "</tr>"
                week = []

        if date.month == 11:
            day_name = date.strftime("%A").lower()
            am_availability = ", ".join(caregiver_calendar.get(day_name, {}).get("am", []))
            pm_availability = ", ".join(caregiver_calendar.get(day_name, {}).get("pm", []))
            cell_content = f"""
                <div class='date-number'>{date.day}</div>
                <div><strong>AM:</strong> {am_availability}</div>
                <div><strong>PM:</strong> {pm_availability}</div>
            """
            week.append(f"<td>{cell_content}</td>")
        else:
            week.append("<td class='other-month'></td>")

        if date.weekday() == 5:
            schedule += "<tr>" + "".join(week) + "</tr>"
            week = []

    if week:
        while len(week) < 7:
            week.append("<td class='other-month'></td>")
        schedule += "<tr>" + "".join(week) + "</tr>"

    schedule += """
        </table>
    </body>
    </html>
    """

    # writes the combined schedule to an HTML file
    with open("Caregiver_Schedule_November_2024.html", "w") as file:
        file.write(schedule)
    print("Generated Caregiver Schedule for November 2024")

#Created by Tanushree
class Pay(Caregiver):
    def __init__(self, caregivers):
        self.caregivers = caregivers

    def calculate_weekly_hours_and_pay(self):
        weekly_summary = {}
        for caregiver in self.caregivers:
            total_hours = 0
            # Each shift (AM or PM) counts as 6 hours
            shift_hours = 6
            for day in caregiver.am.keys():
                if caregiver.am[day] in ["available", "preferred"]:
                    total_hours += shift_hours
                if caregiver.pm[day] in ["available", "preferred"]:
                    total_hours += shift_hours
            caregiver.hours = total_hours
            weekly_pay = total_hours * caregiver.payrate
            weekly_summary[caregiver.name] = {"hours": total_hours, "pay": weekly_pay}
        return weekly_summary


# caregiver object
anne = Caregiver("anne", "443-891-3933", "anne@gmail.com", 20, 10)
anne.set_preferences()
anne.weekly_schedule()

john = Caregiver("john", "301-287-5690", "john@gmail.com", 25, 15)
john.set_preferences()
john.weekly_schedule()


#Created by Aritro
# generates the calendar
caregiver_calendar = Caregiver.generate_calendar()
print("Caregiver Schedule Calendar:")
for day, times in caregiver_calendar.items():
    # displays the schedule for each day
    print(f"{day.capitalize()} - AM: {', '.join(times['am'])} | PM: {', '.join(times['pm'])}")

#Created by Tanushree 
#Calculate weekly pay
caregivers = [anne, john]
pay_calculator = Pay(caregivers)
weekly_summary = pay_calculator.calculate_weekly_hours_and_pay()

#Display weekly summary
print("\nWeekly Pay Summary:")
for caregiver, summary in weekly_summary.items():
    print(f"{caregiver}: {summary['hours']} hours worked, ${summary['pay']} earned")
    
#Created by Aritro    
generate_big_calendar(caregiver_calendar)
```
